```
import os
os.environ['GROQ_API_KEY'] = 'YOUR_GROQ_API_KEY'
```

```
from langchain_groq import ChatGroq
from langchain.agents import initialize_agent, AgentType
from langchain.tools import Tool

from langchain_community.tools import DuckDuckGoSearchRun
from langchain_community.utilities.wikipedia import WikipediaAPIWrapper
```

```
# LLM
llm = ChatGroq(model='llama3-70b-8192', temperature=0.3)
```

```
# Tools
search = DuckDuckGoSearchRun()
wiki = WikipediaAPIWrapper()

def wiki_search(query):
    return wiki.run(query)

tools = [
    Tool(name='WebSearch', func=search.run, description='Search internet'),
    Tool(name='Wikipedia', func=wiki_search, description='Factual knowledge')
]
```

```
# Agent
agent = initialize_agent(
    tools,
    llm,
    agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION,
    verbose=True
)
```

```
def analyze(topic, raw):
    prompt = f"""
    Analyze the research on {topic}.

    {raw}

    Extract:
    - Key insights
    - Trends
    - Important facts
    """
    return llm.invoke(prompt).content
```

```
def generate_report(topic, analysis):
    prompt = f"""
    Create a structured report.

    Topic: {topic}

    {analysis}

    FORMAT:

    COVER PAGE
    Title: {topic}
    Author: Autonomous Agent
    Date: 2026

    Introduction
    Key Findings
    Challenges
    Future Scope
    Conclusion
    """
    return llm.invoke(prompt).content
```

```
def run(topic):
    print('🔍 Researching...\n')
    research = agent.run(f"Research deeply about {topic}")

    print('\n🧠 Analyzing...\n')
    analysis = analyze(topic, research)

    print('\n📄 Generating Report...\n')
    report = generate_report(topic, analysis)

    return report
```

```
topic = 'Impact of AI in Healthcare'
final_report = run(topic)

print('\n✅ FINAL REPORT:\n')
print(final_report)
```
